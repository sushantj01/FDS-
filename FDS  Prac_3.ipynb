```python

import seaborn as sns
import matplotlib.pyplot as plt
 
# load the tips dataset present by default in seaborn
cars = sns.load_dataset('Toyota')
sns.set_style('whitegrid')
 
# make a countplot
sns.countplot(x ='Age', data = cars)
```
