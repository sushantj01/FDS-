# Bar plot

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. 

```python
wine_data =  pd.read_csv("winemag-data-130k-v2.csv")
wine_data
```

```python
wine_data.head()
```

```python
fig,ax=plt.subplots()
data=wine_data["points"].value_counts()

points= data.index
frequency = data.values

ax.bar(points,frequency)

ax.set_title("wine review scored")
ax.set_xlabel("points")
ax.set_ylabel("frequegfhhnfybh")

```

```python
wine_data['points'].value_counts().sort_index().plot.bar()
```

```python
# countris on x-axis an frequency on the y-axis

wine_data.groupby('country').price.mean().sort_values(ascending=False)[:20].plot.bar()
```

```python
#bar plot other charcters we can add

import numpy as np
import matplotlib.pyplot as plt
 
objects = {"Python", "C++", "Java", "Perl", "Scala", "List"}
y_pos = np.arange(len(objects))
performance = [10, 8, 6, 4, 2, 1]
 
plt.bar(y_pos, performance, width=0.5, align="center", alpha=0.5, color=['r', 'r', 'g', 'g', 'b', 'b'])
plt.xticks(y_pos, objects, rotation="horizontal")
plt.ylabel("Usage")
plt.xlabel("Programming Languages")
plt.title("Programming Language Usage")
```

```python
import matplotlib.pyplot as plt
plt.figure (figsize  = (12,7))

countries= ["UAE","USA","INDIA","CHINA","JAPAN","UK"]
total_deaths= [50000,37312,100000,45789,3488,9000]

plt.bar(countries,total_deaths,width=0.9,align="center",color="orange",edgecolor="red")

i= 1.0
j= 2000

for  i in range(len(countries)):
    plt.annotate(total_deaths[i],(-0.1 + i , total_deaths[i]+j))

plt.legend(labels= ['total_deaths'])
plt.ylabel('deaths')
plt.xlabel('countries')

plt.title('total deaths')
plt.show()
plt.savefig('country_deaths.png')
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

labels=['civil','electrical','mechanical','chemical']
sizes = [15,20,16,20]


fig,ax= plt.subplots()
explode= (0.1,0.1,0.8,0.1)

ax.pie(sizes,
       labels = labels,
       autopct= '%1.1f%%',
      explode=explode)

ax.axis('equal')
ax.set_title('engineering diciplines')

plt.show()


```
