```python
import pandas as pd
names=['sushant','raj','ram','arvind','akash']
grades=   [98,67,78,35,89]
bsdegrees=[1,2,3,4,5]
msdegrees=[0,1,2,3,4]
phddegrees=[2,3,5,6,0]

Degrees = zip(names,grades,bsdegrees,msdegrees,phddegrees)
columns = ['Name','Grades','Bs','Ms','Phd']

df=pd.DataFrame(data=Degrees,columns=columns)
df
```

```python
import pandas as pd
names=['sushant','raj','ram','arvind','akash']
grades=   [98,67,78,35,89]
GradeList =zip(names,grades)
df=pd.DataFrame(data=GradeList,columns=['Name','Grades'])

writer=pd.ExcelWriter('dataframe.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name="Sheet1")
writer.save()

```
